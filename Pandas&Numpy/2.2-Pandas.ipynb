```python
# 1. Count distance to previous zeroes
# be [1, 2, 0, 1, 2, 3, 4, 0, 1, 2] for [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]
#
import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
ans=[]
ind=1
for index,row in df.iterrows():
    if (row["X"]==0):
        ans.append(0)
        ind=1
    else:
        ans.append(ind)
        ind = ind+1
    
df["Y"]= ans
df
```

2) Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers.

```python
dates = pd.date_range(start="01/01/2015", end="31/12/2015",  freq='B')
numbers= np.random.randint(1,100,dates.size)
df = pd.DataFrame(numbers,index=dates)
df.columns = ["number"]
df
```

3) Find the sum of the values in s for every Wednesday (0-> monday)

```python
new_df = df[df["number"].index.dayofweek==3]
new_df.sum()
```

4) Average For each calendar month

```python
#g = dates.groupby(pd.Grouper(freq="M"))
df.resample("M").mean()
```

5) For each group of four consecutive calendar months in s, find the date on which the highest value occurred.

```python
#df.resample("4M").max()
df.groupby(pd.Grouper(freq='4M')).max().head(3)
```
